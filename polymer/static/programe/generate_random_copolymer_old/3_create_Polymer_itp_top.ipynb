```python

sobtop_directory = "/path/to/example/static/sobtop"
```

```python



def run_sobtop(mol2_path, chg_path, top_path, itp_path):
    
    
    commands = f'7\n10\n{chg_path}\n0\n1\n2\n4\n{top_path}\n{itp_path}\n0\n'
    
    
    subprocess.run(['cd', sobtop_directory], shell=True)
    
    
    
    process = subprocess.Popen(
        ['./sobtop', mol2_path],
        cwd=sobtop_directory,
        stdin=subprocess.PIPE,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    
    stdout, stderr = process.communicate(commands)

    
    print(stdout)
    
    
    if stderr:
        print(stderr)
```

```python

def create_polymer_itp_top(copolymer_name):
    for filename in os.listdir('.'):
        
        base_filename = os.path.splitext(filename)[0]
        if base_filename == copolymer_name:
            print(base_filename)
            
            mol2_path = os.path.abspath(base_filename + '.mol2')
            chg_path = os.path.abspath(base_filename + '.chg')
            top_path = os.path.abspath(base_filename + '.top')
            itp_path = os.path.abspath(base_filename + '.itp')
            
            run_sobtop(mol2_path, chg_path, top_path, itp_path)
```

```python
def main():
    
    df = pd.read_excel('System_random_copolymer.xlsx')
    copolymer_name = str(df['copolymer_name'][0])
    print(f"--------{copolymer_name}-------{copolymer_name}----------{copolymer_name}--------{copolymer_name}---------{copolymer_name}-------{copolymer_name}----------{copolymer_name}--------{copolymer_name}--------")
    
    create_polymer_itp_top(copolymer_name)
    print(f"--------{copolymer_name}-------{copolymer_name}----------{copolymer_name}--------{copolymer_name}---------{copolymer_name}-------{copolymer_name}----------{copolymer_name}--------{copolymer_name}--------")
```

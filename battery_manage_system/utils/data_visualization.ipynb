```python
import pickle
import numpy as np
```

```python
pkl_path = "/path/to/example/media/bms/pkl_file/battery_data_full_0225_modified_sort_by_cathode.pkl"
with open(pkl_path,"rb") as f:
    data = pickle.load(f)
    print(data["b1"]["info"])
```

```python
import matplotlib.pyplot as plt
```

```python
bat_keys = list(data.keys())
for bat_key in bat_keys:
    print(data[bat_key]["info"]["max_index_below_threshold"])
```

```python
i=0
miaow=[]

for bat_key in bat_keys:
    discharge_capacity_list=[]
    cycle_num = len(data[bat_key]["cycledata"].keys())
    for cycle in range(26,cycle_num+1):
        discharge_capacity = max(data[bat_key]["cycledata"][str(cycle)]["Qdlin"])
        discharge_capacity_list.append(discharge_capacity)
    threhold = 0.8*max(discharge_capacity_list[:10])
    indices = [i for i, x in enumerate(discharge_capacity_list) if x < threhold]
    indices = np.array(indices)
    discharge_capacity_list = np.array(discharge_capacity_list)
    if bat_key=="b2": 
        plt.plot(discharge_capacity_list)
        plt.show()
    print(discharge_capacity_list.argmax())
    if discharge_capacity_list.argmax() > 500:
        print(bat_key)
    if indices.size > 0:
        
        
        i+=1
        max_index_below_threshold=indices.min()+26
    else:
        max_index_below_threshold=cycle_num
    print(str(max_index_below_threshold) in list(data[bat_key]["cycledata"].keys()))
    print(max_index_below_threshold)
    
    miaow.append(max_index_below_threshold)
print(i)
```

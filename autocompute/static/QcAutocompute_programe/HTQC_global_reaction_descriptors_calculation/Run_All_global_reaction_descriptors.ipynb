```python


notebooks_to_run = ['reaction_1_generate_Gaussian_inputfile.ipynb', 'reaction_2_opt+freq_calculation.ipynb'
                   , 'reaction_3_opt+freq_failure_correction.ipynb', 'reaction_4_opt+freq_imaginary_frequencies.ipynb', 'reaction_5_energy_calculation.ipynb'
                   , 'reaction_6_Extracting_energy_and_free_energy_corrections.ipynb']
```

```python

for notebook_filename in notebooks_to_run:
    if not os.path.exists(notebook_filename):
        print(f"File {notebook_filename} not found, breaking...")
        break

    with open(notebook_filename) as f:
        nb = nbformat.read(f, as_version=4)
        
        
        
        ep = ExecutePreprocessor(timeout=None, kernel_name='python3')

        
        try:
            print(f"Executing {notebook_filename}...")
            ep.preprocess(nb, {'metadata': {'path': '.'}})
            print(f"Finished executing {notebook_filename}.")

            
            
            with open(notebook_filename, 'wt', encoding='utf-8') as f:
                nbformat.write(nb, f)
        except Exception as e:
            print(f"Error executing {notebook_filename}: {e}")
            break
```

```python

def get_atom_count_from_mol2(mol2_path):
    
    
    atom_count_regex = re.compile(r'^\s*\d+\s+\d+')
    with open(mol2_path, 'r') as file:
        for line in file:
            
            if atom_count_regex.match(line):
                
                atom_count = int(line.split()[0])
                return atom_count
    return 0  
```

```python
import os, re  

def parse_bondmatrix_txt(txt_or_path):
    
    
    
    if os.path.exists(txt_or_path):  
        with open(txt_or_path, 'r', encoding='utf-8', errors='ignore') as f:  
            text = f.read()  
    else:
        text = str(txt_or_path)  

    
    
    header_regex = re.compile(r'^\s*(\d+(?:\s+\d+)+)\s*$')  
    
    float_pat = r'[+\-]?\d+(?:\.\d*)?(?:[eEdD][+\-]?\d+)?'  
    row_regex = re.compile(
        rf'^\s*(\d+)\s+({float_pat}(?:\s+{float_pat})+)\s*$'  
    )

    
    
    def should_skip(line):
        s = line.strip()  
        if not s:  
            return True
        if '*' in s and set(s) <= set('* '):  
            return True
        low = s.lower()  
        if 'bond order matrix' in low:  
            return True
        if low.startswith('note'):  
            return True
        
        
        
            
        return False  

    
    rows = []  
    
    assigned = {}                     
    current_cols = None              
    max_row = 0                      
    max_col = 0                      

    for line in text.splitlines():   
        
        m_head = header_regex.match(line)  
        if m_head:
            
            cols = [int(x) for x in m_head.group(1).split()]   
            current_cols = cols                                
            if cols:
                max_col = max(max_col, max(cols))              
            continue                                           

        
        if should_skip(line):
            continue

        
        m = row_regex.match(line)                              
        if not (m and current_cols):                           
            continue

        row_idx_1based = int(m.group(1))                       
        floats_str = m.group(2).replace('D','E').replace('d','E')  
        vals = [float(tok) for tok in floats_str.split()]      

        
        kmax = min(len(vals), len(current_cols))

        
        for k in range(kmax):
            i = row_idx_1based - 1                             
            j = current_cols[k] - 1                            
            assigned[(i, j)] = vals[k]                         

        
        if row_idx_1based > max_row:
            max_row = row_idx_1based

    
    n = max(max_row, max_col)                               

    rows = [[0.0] * n for _ in range(n)]                    
    for (i, j), v in assigned.items():                      
        rows[i][j] = v

    
    for i in range(n):
        if rows[i][i] == 0.0:                               
            offsum = sum(rows[i][j] for j in range(n) if j != i)
            rows[i][i] = offsum

    
    
        
    n = len(rows)  
    if any(len(r) != n for r in rows):  
        raise ValueError("Public message removed for release.")

    
    
    tol = 1e-4  
    for i, r in enumerate(rows):  
        diag = r[i]  
        offsum = sum(v for j, v in enumerate(r) if j != i)  
        if abs(diag - offsum) > max(tol, 1e-6 * max(1.0, abs(offsum))):  
            
            
            pass

    return rows  
```

```python


import math                           
import torch                          

def bondmatrix_to_mol2_bonds(bondmatrix, threshold=0.5):
    
    
    try:
        n = len(bondmatrix)                     
        mat = [list(row) for row in bondmatrix] 
        assert all(len(row) == n for row in mat), "Public message removed for release."  
    except Exception as e:
        raise ValueError("Public message removed for release.") from e

    
    def _round_to_mol2_type(x: float) -> int:
        
        if not (x == x):                       
            return 1
        r = int(math.floor(x + 0.5))          
        if r < 1: r = 1                        
        if r > 3: r = 3                        
        return r

    
    lines = ["@<TRIPOS>BOND"]                  
    bond_id = 1                                
    row_src, row_dst = [], []                  

    for i in range(n):                         
        for j in range(i + 1, n):              
            try:
                val = float(mat[i][j])         
            except Exception:
                continue                       
            if val >= threshold:               
                btype = _round_to_mol2_type(val)                 
                lines.append(f"{bond_id:>6d} {i+1:>6d} {j+1:>6d} {btype:>4d}")  
                row_src.append(i)              
                row_dst.append(j)              
                bond_id += 1                   

    num_bonds = bond_id - 1                    

    
    if num_bonds == 0:                         
        edge_index = torch.empty((2, 0), dtype=torch.long)
    else:
        edge_index = torch.tensor([row_src, row_dst], dtype=torch.long)

    
    return "\n".join(lines), num_bonds, edge_index
```

```python

import os  
import re  

def patch_mol2_bonds(name: str, bond_info: str, bond_number: int) -> None:
    
    
    mol2_path = f"{name}.mol2"  
    if not os.path.exists(mol2_path):  
        raise FileNotFoundError("Public message removed for release.")

    with open(mol2_path, "r", encoding="utf-8", errors="ignore") as f:  
        text = f.read()

    
    
    bond_block_regex = re.compile(r"(?is)^@<TRIPOS>BOND\s*.*\Z", flags=re.MULTILINE)
    
    if bond_block_regex.search(text):  
        
        replacement = bond_info.strip() + "\n"
        text = bond_block_regex.sub(replacement, text)  
    else:
        
        text = text.rstrip() + "\n\n" + bond_info.strip() + "\n"

    
    
    mol_head = re.search(r"(?im)^@<TRIPOS>MOLECULE\s*$", text)  
    atom_head = re.search(r"(?im)^@<TRIPOS>ATOM\s*$", text)     
    if not mol_head or not atom_head or atom_head.start() < mol_head.end():
        
        with open(mol2_path, "w", encoding="utf-8") as f:
            f.write(text)
        return

    
    mol_section = text[mol_head.end():atom_head.start()]  
    mol_lines = mol_section.splitlines(keepends=True)     

    
    count_line_re = re.compile(
        ,
        re.VERBOSE
    )

    
    replaced = False  
    for i, ln in enumerate(mol_lines):
        m = count_line_re.match(ln)  
        if m:
            
            new_ln = (
                f"{m.group(1)}{m.group(2)}{m.group(3)}{bond_number}"
                f"{m.group(5)}{m.group(6)}{m.group(7)}{m.group(8)}"
                f"{m.group(9)}{m.group(10)}{m.group(11)}"
            )
            mol_lines[i] = new_ln  
            replaced = True        
            break                  

    
    if replaced:
        new_mol_section = "".join(mol_lines)  
        
        text = text[:mol_head.end()] + new_mol_section + text[atom_head.start():]

    
    with open(mol2_path, "w", encoding="utf-8") as f:  
        f.write(text)  
```

```python

import torch

def mol2_bonds_to_edge_index(mol2_path: str) -> torch.LongTensor:
    
    
    try:
        with open(mol2_path, "r", encoding="utf-8") as f:
            lines = f.readlines()
    except UnicodeDecodeError:
        
        with open(mol2_path, "r", encoding="latin-1") as f:
            lines = f.readlines()

    
    start = None
    for idx, line in enumerate(lines):
        if line.strip().upper().startswith("@<TRIPOS>BOND"):
            start = idx + 1              
            break
    if start is None:
        raise ValueError("Public message removed for release.")

    
    end = len(lines)
    for idx in range(start, len(lines)):
        if lines[idx].strip().startswith("@<TRIPOS>"):
            end = idx
            break

    
    edges = set()  
    for raw in lines[start:end]:
        s = raw.strip()
        if not s or s.startswith("#"):     
            continue
        toks = s.split()                   
        if len(toks) < 3:
            
            continue
        
        try:
            a1 = int(toks[1])  
            a2 = int(toks[2])  
        except ValueError:
            
            continue

        
        i0, j0 = a1 - 1, a2 - 1
        if i0 == j0:
            continue  
        if i0 > j0:
            i0, j0 = j0, i0
        edges.add((i0, j0))

    
    if not edges:
        return torch.empty((2, 0), dtype=torch.long)

    edge_list = sorted(edges, key=lambda x: (x[0], x[1]))  
    src = [e[0] for e in edge_list]
    dst = [e[1] for e in edge_list]
    edge_index = torch.tensor([src, dst], dtype=torch.long)

    return edge_index
```

```python
import torch  

def is_edge_index_equal_undirected(edge_index_mol2: torch.LongTensor,
                                   edge_index_Mayer: torch.LongTensor) -> bool:
    

    def _canonicalize_undirected(ei: torch.LongTensor) -> torch.LongTensor:
        
        if ei is None:  
            return torch.empty((2, 0), dtype=torch.long)  
        if ei.ndim != 2 or ei.shape[0] != 2:  
            raise ValueError("Public message removed for release.")
        if ei.numel() == 0:  
            return torch.empty((2, 0), dtype=torch.long)

        ei = ei.detach().to("cpu").long()  

        a = ei[0]                          
        b = ei[1]                          
        mask = (a != b)                    
        a = a[mask]                        
        b = b[mask]                        

        if a.numel() == 0:                 
            return torch.empty((2, 0), dtype=torch.long)

        lo = torch.minimum(a, b)           
        hi = torch.maximum(a, b)           
        edges = torch.stack([lo, hi], dim=1)  

        edges = torch.unique(edges, dim=0)    

        
        keys = (edges[:, 0].to(torch.int64) << 32) | edges[:, 1].to(torch.int64)  
        order = torch.argsort(keys)            
        edges = edges[order]                   

        return edges.t().contiguous()          

    A = _canonicalize_undirected(edge_index_mol2)  
    B = _canonicalize_undirected(edge_index_Mayer)   

    if A.shape[1] != B.shape[1]:              
        return False

    return bool(torch.equal(A, B))            
```

```python


def read_and_modify_itp_files(directory):
    atomtypes_data = ''
    for filename in os.listdir(directory):
        if filename.endswith('.itp'):
            with open(filename, 'r') as file:
                lines = file.readlines()

            
            for index, line in enumerate(lines):
                if atomtypes_pattern.search(line):
                    start_index = index + 2
                    break
            
            
            atomtypes_content = []
            for line in lines[start_index:]:
                if empty_line_pattern.search(line):
                    break
                atomtypes_content.append(line)
            
            atomtypes_data += ''.join(atomtypes_content)

            
            del lines[start_index - 2:index + len(atomtypes_content) + 2]

            
            with open(filename, 'w') as file:
                file.writelines(lines)

    return atomtypes_data
```

```python

def update_itp_files(excel_path, itp_folder):
    
    df = pd.read_excel(excel_path)

    
    serial_to_name = dict(zip(df['Name'], df['Serial Number'].fillna(0).astype(int)))

    
    os.chdir(itp_folder)

    
    itp_files = glob.glob('*.itp')

    
    for itp_file in itp_files:
        name = os.path.splitext(itp_file)[0]  
        if name in serial_to_name:  
            serial_number = serial_to_name[name]  
            serial_number = str(serial_number)
            with open(itp_file, 'r') as file:
                lines = file.readlines()
            
            with open(itp_file, 'w') as file:
                for line in lines:
                    
                    if re.match(r';\s+Index\s+type\s+residue\s+resname\s+atom\s+cgnr\s+charge\s+mass', line):
                        file.write(line)  
                        continue
                    
                    match = re.match(r'(\s*\d+\s+)(\w+)(\s+\w+\s+\w+\s+)(\w+)(\s+\d+\s+[-+]?\d*\.?\d+\s+[-+]?\d*\.?\d+)', line)
                    
                    if match:
                        type_col = match.group(2)
                        atom_col = match.group(4)
                        new_atom_col = f"{type_col}_{serial_number}"  
                        
                        new_line = f"{match.group(1)}{type_col}{match.group(3)}{new_atom_col}{match.group(5)}\n"
                        file.write(new_line)
                    else:
                        file.write(line)  
        else:
            print(f"Warning: '{name}' not found in Excel mapping. Skipping file '{itp_file}'.")
```

```python

def read_charges_from_itp(excel_path, itp_folder):
    
    
    
    df = pd.read_excel(excel_path)

    
    name_to_serial = dict(zip(df['Name'], df['Serial Number'].fillna(0).astype(int)))

    
    itp_files = glob.glob(os.path.join(itp_folder, '*.itp'))
    
    
    charges_dict = {}
    
    
    for itp_file in itp_files:
        charges_list = []  
        name = os.path.splitext(os.path.basename(itp_file))[0]  
        if name in name_to_serial:  
            with open(itp_file, 'r') as file:
                lines = file.readlines()
                
                
                if name in df['Name'].values:
                    scale_charge = df.loc[df['Name'] == name, 'scale_charge'].iloc[0]
                else:
                    scale_charge = 1
                
                for line in lines:
                    
                    match = re.match(r'(\s*\d+\s*)(\w+)\s+(\d+)\s+(\w+)\s+(\w+_\d+)\s+(\d+)\s+([-+]?\d*\.\d+)\s+([-+]?\d*\.\d+)', line)
                    if match:
                        charge_col = match.group(7)  
                        charges_list.append(float(charge_col))
        
        total_charge = sum(charges_list)  
        charges_dict[name] = total_charge  

        
        
        
        charges_list_scaled = [charge * scale_charge for charge in charges_list]

        
        charges_list_zero = [0 * scale_charge for charge in charges_list]

        
        file_name_without_extension = get_filename_without_extension(itp_file)

        
        new_file_name = f"without_charge_{file_name_without_extension}.itp"
        
        with open(itp_file, 'r') as file:
            lines = file.readlines()

        
        with open(new_file_name, 'w') as file:
            charge_index = 0  
            for line in lines:
                
                match = re.match(r'(\s*\d+\s+\w+\s+\d+\s+\w+\s+\w+_\d+\s+\d+\s+)([-+]?\d*\.\d+)(\s+[-+]?\d*\.\d+)', line)
                if match and charge_index < len(charges_list_scaled):
                    
                    new_charge_str = f"{charges_list_zero[charge_index]: .6f}"
                    new_line = match.group(1) + new_charge_str + match.group(3) + '\n'
                    file.write(new_line)
                    charge_index += 1
                else:
                    
                    file.write(line)
        
        with open(itp_file, 'w') as file:
            charge_index = 0  
            for line in lines:
                
                match = re.match(r'(\s*\d+\s+\w+\s+\d+\s+\w+\s+\w+_\d+\s+\d+\s+)([-+]?\d*\.\d+)(\s+[-+]?\d*\.\d+)', line)
                if match and charge_index < len(charges_list_scaled):
                    
                    new_charge_str = f"{charges_list_scaled[charge_index]: .6f}"
                    new_line = match.group(1) + new_charge_str + match.group(3) + '\n'
                    file.write(new_line)
                    charge_index += 1
                else:
                    
                    file.write(line)

    return charges_dict
```

```python

import re
from itertools import combinations
from pathlib import Path


def build_unique_resname_map(names) -> dict[str, str]:
    

    def _normalize_name(raw_name) -> str:
        
        name = str(raw_name).strip()
        if not name:
            raise ValueError("Public message removed for release.")
        return name

    def _extract_letters(name: str) -> str:
        
        return ''.join(ch for ch in name.upper() if 'A' <= ch <= 'Z')

    def _generate_primary_candidates(letters: str) -> list[str]:
        
        candidates: list[str] = []
        seen_candidates: set[str] = set()

        if len(letters) >= 3:
            for index_triplet in combinations(range(len(letters)), 3):
                candidate = ''.join(letters[idx] for idx in index_triplet)
                if candidate not in seen_candidates:
                    seen_candidates.add(candidate)
                    candidates.append(candidate)
        elif letters:
            candidates.append(letters)

        return candidates

    def _generate_fallback_candidates(letters: str):
        
        if len(letters) >= 2:
            roots = [letters[:2]]
        elif len(letters) == 1:
            roots = [letters]
        else:
            roots = ['X']

        yielded: set[str] = set()
        for root in roots:
            max_digit_width = 3 - len(root)
            for width in range(1, max_digit_width + 1):
                upper = 10 ** width
                for number in range(upper):
                    candidate = f'{root}{number:0{width}d}'
                    if candidate not in yielded:
                        yielded.add(candidate)
                        yield candidate

    normalized_names: list[str] = []
    seen_names: set[str] = set()
    for raw_name in names:
        name = _normalize_name(raw_name)
        if name not in seen_names:
            seen_names.add(name)
            normalized_names.append(name)

    resname_map: dict[str, str] = {}
    used_resnames: dict[str, str] = {}

    for name in normalized_names:
        letters = _extract_letters(name)
        chosen_resname = None

        for candidate in _generate_primary_candidates(letters):
            if 1 <= len(candidate) <= 3 and candidate not in used_resnames:
                chosen_resname = candidate
                break

        if chosen_resname is None:
            for candidate in _generate_fallback_candidates(letters):
                if 1 <= len(candidate) <= 3 and candidate not in used_resnames:
                    chosen_resname = candidate
                    break

        if chosen_resname is None:
            raise ValueError(
                "Public message removed for release."
            )

        resname_map[name] = chosen_resname
        used_resnames[chosen_resname] = name

    return resname_map



def rewrite_pdb_resname_for_targets(df, resname_map, cwd='.') -> dict:
    
    if 'Name' not in df.columns:
        raise KeyError("Public message removed for release.")

    base_dir = Path(cwd).resolve()
    report = {'modified': [], 'skipped': [], 'not_found': []}
    processed_names: set[str] = set()

    for row_index, raw_name in df['Name'].items():
        name = str(raw_name).strip()
        if not name:
            report['skipped'].append({'row_index': row_index, 'name': raw_name, 'reason': 'empty_name'})
            continue
        if name in processed_names:
            continue
        processed_names.add(name)

        if name not in resname_map:
            raise KeyError("Public message removed for release.")

        pdb_path = base_dir / f'{name}.pdb'
        if not pdb_path.is_file():
            report['not_found'].append({'row_index': row_index, 'name': name, 'file': str(pdb_path)})
            continue

        lines = pdb_path.read_text(encoding='utf-8', errors='ignore').splitlines(keepends=True)
        out_lines: list[str] = []
        changed = False
        target_resname = resname_map[name]

        for line in lines:
            record_name = line[:6]
            if record_name not in ('ATOM  ', 'HETATM'):
                out_lines.append(line)
                continue

            has_newline = line.endswith(chr(10))
            body = line[:-1] if has_newline else line
            if len(body) < 20:
                body = body.ljust(20)

            new_body = body[:17] + f'{target_resname:>3}' + body[20:]
            new_line = new_body + (chr(10) if has_newline else '')
            out_lines.append(new_line)
            if new_line != line:
                changed = True

        if changed:
            pdb_path.write_text(''.join(out_lines), encoding='utf-8')
            report['modified'].append({'row_index': row_index, 'name': name, 'file': str(pdb_path), 'new_resname': target_resname})
        else:
            report['skipped'].append({'row_index': row_index, 'name': name, 'file': str(pdb_path), 'reason': 'no_atom_or_hetatm_records_changed'})

    return report



def rewrite_itp_resname_for_targets(df, resname_map, cwd='.') -> dict:
    
    if 'Name' not in df.columns:
        raise KeyError("Public message removed for release.")

    atoms_header_re = re.compile(r'^\s*\[\s*atoms\s*\]\s*$', re.IGNORECASE)
    section_header_re = re.compile(r'^\s*\[.*\]\s*$')
    atom_line_re = re.compile(r'^(\s*\d+\s+\S+\s+\d+\s+)(\S+)(\s+\S+\s+\d+.*)$')

    base_dir = Path(cwd).resolve()
    report = {'modified': [], 'skipped': [], 'not_found': []}
    processed_names: set[str] = set()

    for row_index, raw_name in df['Name'].items():
        name = str(raw_name).strip()
        if not name:
            report['skipped'].append({'row_index': row_index, 'name': raw_name, 'reason': 'empty_name'})
            continue
        if name in processed_names:
            continue
        processed_names.add(name)

        if name not in resname_map:
            raise KeyError("Public message removed for release.")

        target_resname = resname_map[name]
        targets = [base_dir / f'{name}.itp', base_dir / f'without_charge_{name}.itp']

        for itp_path in targets:
            if not itp_path.is_file():
                report['not_found'].append({'row_index': row_index, 'name': name, 'file': str(itp_path)})
                continue

            lines = itp_path.read_text(encoding='utf-8', errors='ignore').splitlines(keepends=True)
            out_lines: list[str] = []
            in_atoms_section = False
            changed = False

            for line in lines:
                if (not in_atoms_section) and atoms_header_re.match(line):
                    in_atoms_section = True
                    out_lines.append(line)
                    continue

                if in_atoms_section and section_header_re.match(line) and (not atoms_header_re.match(line)):
                    in_atoms_section = False
                    out_lines.append(line)
                    continue

                if not in_atoms_section:
                    out_lines.append(line)
                    continue

                stripped = line.lstrip()
                if stripped.startswith(';') or stripped.startswith('#') or stripped.strip() == '':
                    out_lines.append(line)
                    continue

                body, sep, comment = line.partition(';')
                has_newline = line.endswith(chr(10))
                body_no_nl = body.rstrip(chr(10))
                match = atom_line_re.match(body_no_nl)
                if match is None:
                    out_lines.append(line)
                    continue

                new_body = f'{match.group(1)}{target_resname}{match.group(3)}'
                if sep:
                    new_line = f"{new_body} {sep}{comment.rstrip(chr(10))}"
                else:
                    new_line = new_body
                if has_newline:
                    new_line += chr(10)

                out_lines.append(new_line)
                if new_line != line:
                    changed = True

            if changed:
                itp_path.write_text(''.join(out_lines), encoding='utf-8')
                report['modified'].append({'row_index': row_index, 'name': name, 'file': str(itp_path), 'new_resname': target_resname})
            else:
                report['skipped'].append({'row_index': row_index, 'name': name, 'file': str(itp_path), 'reason': 'no_atom_lines_matched'})

    return report
```

```python

def update_net_charge(file_path, polymer_name, charges_dict):
    """
    Update the Net Charge column in an Excel file based on the polymer names and charges dictionary.
    
    :param file_path: str, the path to the Excel file to be updated
    :param polymer_name: list, a list of polymer names to check against
    :param charges_dict: dict, a dictionary with component names as keys and atomic charges as values
    """
    
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print(f"Error reading the Excel file: {e}")
        return

    
    if 'Net Charge' not in df.columns:
        df['Net Charge'] = None

    
    for name, charge in charges_dict.items():
        if name in polymer_name:
            
            rounded_charge = round(charge)

            
            index = df[df['Name'] == name].index

            
            if len(index) == 1:
                df.at[index[0], 'Net Charge'] = rounded_charge
            else:
                print(f"Warning: Multiple or no entries found for {name}. No update performed.")

    
    try:
        df.to_excel(file_path, index=False, engine='openpyxl')
        print(f"Excel file '{file_path}' has been updated successfully.")
    except Exception as e:
        print(f"Error writing to the Excel file: {e}")
```

```python
import os, shutil, subprocess, sys
from datetime import datetime
from pathlib import Path

def check_mpi_env(output_dir: str = None) -> str:
    
    
    if output_dir is None:
        output_dir = f"mpi_check_{datetime.now().strftime('%Y%m%d_%H%M%S_%f')}"
    outdir = Path(output_dir).resolve()
    outdir.mkdir(parents=True, exist_ok=False)

    
    gmx_bin = resolve_gmx()

    
    (outdir / "gmx_path.txt").write_text(f"gmx_bin = {gmx_bin}\n", encoding="utf-8")

    
    mpirun_path = shutil.which("mpirun") or "<not found>"
    try:
        mpirun_ver = subprocess.check_output(
            [mpirun_path, "--version"], text=True, stderr=subprocess.STDOUT
        ) if mpirun_path != "<not found>" else ""
    except Exception as e:
        mpirun_ver = f"<error: {e}>"
    (outdir / "which_mpirun.txt").write_text(
        f"which mpirun: {mpirun_path}\n{mpirun_ver}", encoding="utf-8"
    )

    
    full_ldd = ""
    filt_ldd = ""
    try:
        full_ldd = subprocess.check_output(["ldd", gmx_bin], text=True, stderr=subprocess.STDOUT)
        
        lines = [ln for ln in full_ldd.splitlines()
                 if any(k in ln.lower() for k in ("libmpi", "libfabric", "ucx"))]
        filt_ldd = "\n".join(lines) + ("\n" if lines else "")
    except Exception as e:
        full_ldd = "Public message removed for release."

    (outdir / "ldd_gmx_full.txt").write_text(full_ldd, encoding="utf-8")
    (outdir / "ldd_gmx_filtered.txt").write_text(
        filt_ldd if filt_ldd else "Public message removed for release.", encoding="utf-8"
    )

    
    env_dump = []
    for key in ("PATH", "LD_LIBRARY_PATH", "FI_PROVIDER", "FI_TCP_IFACE", "OMPI_MCA_btl", "OMPI_MCA_btl_tcp_if_include"):
        val = os.environ.get(key, "")
        env_dump.append(f"{key}={val}")
    (outdir / "env_keys.txt").write_text("\n".join(env_dump) + "\n", encoding="utf-8")

    
    try:
        gmx_ver = subprocess.check_output([gmx_bin, "--version"], text=True, stderr=subprocess.STDOUT)
    except Exception as e:
        gmx_ver = "Public message removed for release."
    (outdir / "gmx_version.txt").write_text(gmx_ver, encoding="utf-8")

    print("Public message removed for release.")
    return str(outdir)
```

```python
def insert_molecules(data_list, initial_size=10):
    gmx_bin = resolve_gmx()  
    box_size = initial_size  
    max_time = 30            
    attempt_idx = 1          
    log_dir = "logs_insert"  
    os.makedirs(log_dir, exist_ok=True)  
    created_logs = []        

    while True:  
        output_file = ""     
        success_all = True   

        for i, (name, number) in enumerate(data_list, start=1):  
            if i == 1:
                output_file = f"box{i}.pdb"  
                cmd = [
                    gmx_bin, "insert-molecules",                 
                    "-ci", f"{name}.pdb",                        
                    "-box", str(box_size), str(box_size), str(box_size),  
                    "-nmol", str(number),                        
                    "-try", "200000",                            
                    "-o", output_file                            
                ]
            else:
                input_file = output_file                         
                output_file = f"box{i}.pdb"
                cmd = [
                    gmx_bin, "insert-molecules",
                    "-f", input_file,                            
                    "-ci", f"{name}.pdb",
                    "-nmol", str(number),
                    "-try", "200000",
                    "-o", output_file
                ]

            
            tag = f"{attempt_idx}_{i}_{datetime.now().strftime('%Y%m%d_%H%M%S_%f')}"
            out_path = os.path.join(log_dir, f"process_{tag}.out")  
            err_path = os.path.join(log_dir, f"process_{tag}.err")  
            created_logs.extend([out_path, err_path])               

            
            with open(out_path, "x") as fout, open(err_path, "x") as ferr:
                process = subprocess.Popen(  
                    cmd,                     
                    stdout=fout,             
                    stderr=ferr              
                )
                start_time = time.time()     
                print(f"Started process {process.pid} for {name}.")  

                
                while True:
                    if process.poll() is not None:  
                        break
                    if time.time() - start_time > max_time:  
                        try:
                            process.terminate()  
                            print(f"Process {process.pid} terminated due to timeout.")
                        except Exception:
                            pass
                        time.sleep(1)           
                        if process.poll() is None:
                            process.kill()      
                            print(f"Process {process.pid} forcibly killed.")
                        break
                    time.sleep(0.1)             

            
            if process.returncode == 0:
                print(f"Process {process.pid} completed successfully for {name}.")  
                continue  
            else:
                
                for file in glob.glob("#*"):
                    try:
                        os.remove(file)  
                    except FileNotFoundError:
                        pass
                print(f"Failed to insert {name}. Increasing box size to {box_size + 2} nm and retrying.")
                box_size += 2      
                attempt_idx += 1   
                success_all = False  
                break               

        if success_all:
            
            print(f"Insertion complete. Final box size: {box_size} nm.")
            
            for p in created_logs:
                try:
                    os.remove(p)  
                except FileNotFoundError:
                    pass
            break  

    
    shutil.move(output_file, "final_box.pdb")  
```

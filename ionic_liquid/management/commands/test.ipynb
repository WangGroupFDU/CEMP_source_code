```python
import pandas as pd


df = pd.read_excel('Dimer.xlsx', engine='openpyxl')
```

```python

column_types = df.dtypes

def generate_django_model_code(df, model_name="MyModel"):
    model_code = f"from django.db import models\n\n"
    model_code += f"class {model_name}(models.Model):\n"
    
    dtype_to_field = {
        'object': "models.CharField(max_length=255)",
        'int64': "models.IntegerField()",
        'float64': "models.FloatField()",
        'bool': "models.BooleanField()",
        
    }
    
    for column_name, dtype in df.dtypes.items():
        
        column_name = column_name.replace(" ", "_")
        column_name = column_name.replace("(", "")
        column_name_safe = column_name.replace(")", "")
        field_type = dtype_to_field.get(str(dtype), "models.CharField(max_length=255)")
        model_code += f"    {column_name_safe} = {field_type}\n"
    
    return model_code


df = pd.read_excel('Dimer.xlsx', engine='openpyxl')


model_code = generate_django_model_code(df)
print(model_code)
```
